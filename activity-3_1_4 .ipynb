```python
import pandas as pd 
```

```python
movies_merge = pd.read_excel('movies_merge.xlsx')
ott_merge = pd.read_csv('ott_merge.csv')
```

```python
print(movies_merge.shape)
movies_merge.head()
```

```python
print(ott_merge.shape)
ott_merge.head()
```

```python
print(movies_merge.dtypes)
print(ott_merge.dtypes)
```

```python
movie_columns = list(movies_merge)
movie_columns
```

```python
ott_columns = list(ott_merge)
ott_columns
```

```python
# combine using concat 
movies_ott_concat = pd.concat([movies_merge, ott_merge], axis=0)
movies_ott_concat
```

```python
# combine using merge 

movies_ott_merge = movies_merge.merge(ott_merge, how='outer', on='ID')
movies_ott_merge
```

```python
# combine using append() 
movies_ott_append = movies_merge.append(ott_merge, ignore_index=False)
movies_ott_append
```

```python
print(movies_ott_concat.shape)
print(movies_ott_merge.shape)
print(movies_ott_append.shape)
```

```python
#debug 
# assert new_df.shape[0] == (df_left.shape[0]  + df_right.shape[0])
assert movies_ott_concat.shape[0] == (movies_merge.shape[0] + ott_merge.shape[0])
# no output so code has been exceuted correctly! 
```
