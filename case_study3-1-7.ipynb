```python
import pandas as pd 
movies1 = pd.read_excel('movies_merge.xlsx')
ott1 = pd.read_csv('ott_merge.csv')

print(movies1.shape)
print(movies1.columns)
print(ott1.shape)
print(ott1.columns)
```

```python

movies_ott = pd.merge(movies1, ott1, how='left', on='ID')
movies_ott
```

 How many films from each year (released from 2012 to the present) were watched on Netflix?

```python
q1 = movies_ott.groupby('Year')[['Netflix']].sum().reset_index()

q1[q1['Year'] >= 2012]
```

What is the average runtime of movies released each year?

```python
q2 = movies_ott.groupby('Year')['Runtime'].agg('mean').reset_index()
q2[q2['Year'] >= 2012]
```

What are the best and worst reviews that movies released each year received on Rotten Tomatoes?

```python
q3 = movies_ott.groupby('Year')[['Rotten Tomatoes']].agg(['min', 'max']).reset_index()
q3[q3['Year'] >= 2012]
```
