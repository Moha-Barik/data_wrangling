# NOTE: in almost all cases, groupby() means to follow a process of splitting data into groups according to some criteria. 

```python
# create a new column 'Sales total' by multiplying quantitiy by unitprice. 
transactions_description_country['Sales_total'] = transactions_description_country['Quantity'] * transactions_description_country['UnitPrice']
transactions_description_country
```

```python
# group the sum of TotalSales by country
transactions_description_country.groupby("Country")[['Sales_total']].sum()
```

```python
# sort by country the total_sale 
# the \ indicates a line break without interrupting the code snippet
transactions_description_country.groupby('Country')\
[['Sales_total']].sum().sort_values\
('Sales_total', ascending=False)
```

```python
# total and mean sales by country
transactions_description_country.groupby("Country")[["Sales_total"]] \
.agg(["sum", "mean"])
```

```python
# sort total and average sales by country
transactions_description_country.groupby("Country")[["Sales_total"]] \
.agg(["sum", "mean"]) \
.sort_values([("Sales_total", "sum")], ascending=False)
```

```python
# sort total and average sales by product type 
transactions_description_country.groupby("Description")[["Sales_total"]] \
.agg(["sum", "mean"]) \
.sort_values([("Sales_total", "sum")], ascending=False)
```
